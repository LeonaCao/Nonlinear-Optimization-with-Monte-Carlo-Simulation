## Nonlinear Optimization Model

```
# create a covariance table for the return_data dataframe
return_cov = return_data.cov()      # covariance matrix of the dataframe
print('Covariance Matrix:')
print(return_cov)
print('\n')

# calculate averages of the stocks - coefficients of the objective function
avg_return = return_data.mean()
print('Average Return:')
print(avg_return)
```

```
# set up the model
from pyomo.environ import *
m = ConcreteModel()

# define the variable for each stock
m.aptv = Var(within=NonNegativeReals, bounds=(0,1))     # proportion of stocks allocated in APTV
m.azo = Var(within=NonNegativeReals, bounds=(0,1))     # proportion of stocks allocated in AZO
m.cmg = Var(within=NonNegativeReals, bounds=(0,1))     # proportion of stocks allocated in CMG
m.dltr = Var(within=NonNegativeReals, bounds=(0,1))     # proportion of stocks allocated in DLTR
m.nke = Var(within=NonNegativeReals, bounds=(0,1))     # proportion of stocks allocated in NKE
m.nvr = Var(within=NonNegativeReals, bounds=(0,1))     # proportion of stocks allocated in NVR
m.tif = Var(within=NonNegativeReals, bounds=(0,1))     # proportion of stocks allocated in TIF
m.tpr = Var(within=NonNegativeReals, bounds=(0,1))     # proportion of stocks allocated in TPR
m.uaa = Var(within=NonNegativeReals, bounds=(0,1))     # proportion of stocks allocated in UAA
m.whr = Var(within=NonNegativeReals, bounds=(0,1))     # proportion of stocks allocated in WHR
```

```
# declare the objective function - sum of returns from all stocks
m.objective = Objective(expr =               
                        m.aptv*avg_return[0] +      # returns obtained from stocks allocated to APTV
                        m.azo*avg_return[1] +       # returns obtained from stocks allocated to AZO
                        m.cmg*avg_return[2] +      # returns obtained from stocks allocated to CMG
                        m.dltr*avg_return[3] +       # returns obtained from stocks allocated to DLTR
                        m.nke*avg_return[4] +      # returns obtained from stocks allocated to NKE
                        m.nvr*avg_return[5] +      # returns obtained from stocks allocated to NVR
                        m.tif*avg_return[6] +      # returns obtained from stocks allocated to TIF
                        m.tpr*avg_return[7] +      # returns obtained from stocks allocated to TPR
                        m.uaa*avg_return[8] +      # returns obtained from stocks allocated to UAA
                        m.whr*avg_return[9],      # returns obtained from stocks allocated to WHR
                        sense=maximize)     # maximization of objective function
```

```
# declare the constraints
## 1. sum of all porportions = 1
m.sum_proportions = Constraint(expr = m.aptv + m.azo + m.cmg + m.dltr + m.nke + m.nvr + m.tif + m.tpr + m.uaa + m.whr == 1)

## 2. minimum return = 0.2%
m.return_floor = Constraint(expr = m.objective >= 0.002)
```

```
## 3. create calculations table for calculating the risk
def calc_risk():
  # list of all our investment allocations (proportion invested in each stock)
  # updated as the model is cranking
  proportions = [ m.aptv, m.azo, m.cmg, m.dltr, m.nke, m.nvr, m.tif, m.tpr, m.uaa, m.whr]
  risk_matrix = []
  
  # iterate across the covariance matrix
  # 'risk elements' = cov(A,B)*prop(A)*prop(B)
  # since we have 10 stocks, there are 10*10 = 100 elements
  # sum these 100 elements up at the end, and we get the risk
  for index, col in enumerate(return_cov.columns):
    i = 0
    risk = []
    for row in return_cov[col]:
      risk.append(proportions[index]*row*proportions[i])
      i = i + 1
    risk_matrix.append(risk)
  risk_matrix = pd.DataFrame(risk_matrix)
  return risk_matrix.sum().sum()

# a sequence of risk options
risk_limits = np.around(np.arange(0.00003, 0.00033, 0.000001), 6) # round them to ensure the precisions are 6 decimal places
risk_limits
```

```
# update the risk contraint for each limit and then solve the problem
param_analysis = {} # key = risk, value = stock allocations 
returns = {} # key = risk, value = return
for r in risk_limits:
  m.total_risk = Constraint(expr = calc_risk() <= r)
  # run solver
  result = SolverFactory('ipopt', executable=ipopt_executable).solve(m).write()
  # store our allocation proportions
  param_analysis[r] = [m.aptv(), m.azo(), m.cmg(), m.dltr(), m.nke(), m.nvr(), m.tif(), m.tpr(), m.uaa(), m.whr()]
  # store our returns
  returns[r] = m.aptv()*avg_return[0] + m.azo()*avg_return[1] + m.cmg()*avg_return[2] + m.dltr()*avg_return[3] + m.nke()*avg_return[4] + m.nvr()*avg_return[5] + m.tif()*avg_return[6] + m.tpr()*avg_return[7] + m.uaa()*avg_return[8] + m.whr()*avg_return[9]
```

### Parameter Analysis

```
# generate a dataframe for proportions of the portfolio for each risk limit
param_analysis = pd.DataFrame.from_dict(param_analysis, orient='index')
param_analysis.columns = ['APTV','AZO','CMG','DLTR','NKE','NVR','TIF','TPR','UAA','WHR']    # column names are respective stock abbreviations 
param_analysis
```

```
# plot the dataframe to get stock allocation over various risk values
param_analysis.plot()
plt.yscale('linear')    # specify scale of allocation proportion
plt.xlabel('Risk level')      # name x axis
plt.ylabel('Proportion of stock allocated')     # name y axis
plt.title('Optimal Stock Allocation for Different Risk Levels')       # name the plot
plt.show()
```

```
# subset the risk and return (defined as 'reward')
risk = list(returns.keys()) # coerce dict_keys to a list
print(risk)
reward = list(returns.values()) # coerce dict_values to a list
print(reward)
```

```
# plot the efficient frontier
plt.figure(figsize=(6.5, 4.5))      # set figure size
plot(risk, reward, '-.', color='green')
title('The Efficient Frontier')     # name the plot
xlabel('Risk')    # name the x axis
ylabel('Reward (Return)')     # name the y axis
plt.show()
```

## Allocating $1\:Million\:USD$

In the graph above, it can be seen that for a higher value of risk, the plot starts to flatten, indicating that a risk above approx. $0.0025$ is not going to yield any additional returns. 

To select a risk level that would provide an acceptable allocation of stocks, we looked at how much returns the risk level can be mapped to, as well as the variety of stocks maintained at that risk level. Looking at the Efficient Frontier plot and the Optimal Stock Allocation for different risk levels, we decided on moving ahead with a risk level of $0.00015$, as the slope of efficieny frontier starts to reduce after this risk level, whereas the returns earned at this risk level is above average returns earned. Additionally, at this risk level, the investment is not concentrated on a single stock, but is spread across numerous stocks.
